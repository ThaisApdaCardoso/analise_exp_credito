### Query 4 – Faixa Salarial (Barras Horizontais)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df4 = pd.read_csv('M8-query4-qtd_faixa_salarial.csv')

plt.figure(figsize=(10, 6))
sns.barplot(data=df4, x='total', y='salario_anual', palette='Blues_d')
plt.title('Distribuição por Faixa Salarial')
plt.xlabel('Total de Clientes')
plt.ylabel('Faixa Salarial')
plt.tight_layout()
plt.show()
```

### Query 5 – Distribuição por Sexo (Gráfico de Pizza)

```python
df5 = pd.read_csv('M8-query5-distrib_sexo.csv')

plt.figure(figsize=(6,6))
plt.pie(df5['total'], labels=df5['sexo'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribuição por Sexo')
plt.axis('equal')
plt.show()
```

### Query 6 – Top 10 Limites de Crédito

```python
df6 = pd.read_csv('M8-query6-top10_limites_grupo.csv')
df6['grupo'] = df6['escolaridade'] + ' | ' + df6['tipo_cartao'] + ' | ' + df6['sexo']

plt.figure(figsize=(12, 6))
sns.barplot(data=df6, x='limite_max', y='grupo', palette='Greens_d')
plt.title('Top 10 Limites de Crédito por Perfil')
plt.xlabel('Limite de Crédito')
plt.ylabel('Perfil')
plt.tight_layout()
plt.show()
```

### Query 7 – Valor Transacionado por Sexo

```python
df7 = pd.read_csv('M8-query7-maior_media_valor_sexo.csv')

df7_melt = df7.melt(id_vars='sexo', value_vars=['maior_valor', 'media_valor'],
                    var_name='tipo_valor', value_name='valor')

plt.figure(figsize=(8, 6))
sns.barplot(data=df7_melt, x='sexo', y='valor', hue='tipo_valor', palette='Set2')
plt.title('Valores Transacionados por Sexo')
plt.ylabel('R$')
plt.tight_layout()
plt.show()
```

### Query 8 – Médias por Sexo e Faixa Salarial

```python
df8 = pd.read_csv('M8-query8-medias_sexo_salarial.csv')

plt.figure(figsize=(12, 6))
sns.barplot(data=df8, x='salario_anual', y='media_transacoes', hue='sexo', palette='coolwarm')
plt.title('Média de Transações por Sexo e Faixa Salarial')
plt.xlabel('Faixa Salarial')
plt.ylabel('Média de Transações')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
